```python
# Extract the names of the selected crops and their percentages
selected_crops_data = []
for crop, percentage in selected_crops.items():
    cleaned_crop_name = crop.strip()
    selected_crops_data.append({'crop': cleaned_crop_name, 'percentage': percentage})

# Create a DataFrame from the selected crops data
selected_crops_data = pd.DataFrame(selected_crops_data)

# Define a function to assess the crop selection
def assess_crop_selection(country_data, selected_crops_data, crop_data):
    matching_criteria = []  # To store the matching criteria

    for index, row in selected_crops_data.iterrows():
        selected_crop = row['crop']
        area_percentage = row['percentage']

        # Clean the crop name by stripping leading/trailing spaces
        cleaned_crop_name = selected_crop.strip()

        # Check if the crop name exists in the crop data
        if cleaned_crop_name in crop_data['Name '].values:
            crop_info = crop_data[crop_data['Name '] == cleaned_crop_name].iloc[0]
            
            # Check if the country's average temperature falls within the crop's temperature range
            country_avg_temp = country_data['Avg. Temperature C'].iloc[0]
            crop_temp_range = crop_info['Temperature (oC)']
            crop_temp_min, crop_temp_max = map(int, crop_temp_range.split(" to "))
            if crop_temp_min <= country_avg_temp <= crop_temp_max:
                matching_criteria.append(f"Average Temperature for {cleaned_crop_name}")

            # Check if the country's average precipitation falls within the crop's precipitation range
            country_avg_precip = country_data['Avg. Precipitation (mm)'].iloc[0]
            crop_precip_range = crop_info['Precipitation (mm)']
            crop_precip_min, crop_precip_max = map(int, crop_precip_range.split(" to "))
            if crop_precip_min <= country_avg_precip <= crop_precip_max:
                matching_criteria.append(f"Average Precipitation for {cleaned_crop_name}")

            # Check if the country's soil fertility matches the crop's requirement (high)
            country_soil_fertility = country_data['Soil_fertility (class)'].iloc[0]
            if country_soil_fertility.lower() == 'high':
                matching_criteria.append(f"Soil Fertility for {cleaned_crop_name}")

            # Check if the country's agro-ecological zone matches the crop's climate zone
            # Modify the column names as needed
            country_agro_ecological_zone = country_data['Agro-ecological_zone_class'].iloc[0]
            crop_climate_zone = crop_info['Climate zone']
            if crop_climate_zone in country_agro_ecological_zone:
                matching_criteria.append(f"Agro-ecological Zone and Climate Zone for {cleaned_crop_name}")

    return matching_criteria

# Define your country_data_selected here

# Assess the crop selection
matching_criteria = assess_crop_selection(country_data_selected, selected_crops_data, crop_data)

# Determine if it's a good or poor selection based on the score threshold
score_threshold = 3  # Adjust this threshold as needed
if len(matching_criteria) >= score_threshold:
    print("\nGood selection of crops! The following criteria match:")
    for criteria in matching_criteria:
        print(f"- {criteria}")
else:
    print("\nPoor selection of crops. The matching criteria are:")
    for criteria in matching_criteria:
        print(f"- {criteria}")
```

```python
# Create a dictionary to store the yield per ha for each crop
crop_yield_per_ha = {
    'Cotton': 10,  # Replace with the actual yield per ha for Cotton
    'Maize': 8,    # Replace with the actual yield per ha for Maize
    'Coffee': 12,  # Replace with the actual yield per ha for Coffee
    'Rice': 9,     # Replace with the actual yield per ha for Rice
    'Wheat': 7     # Replace with the actual yield per ha for Wheat
}

# Initialize a dictionary to store the attainable maximum yield and other information for each crop
attainable_info = {}

# Conversion factor from km² to hectares
km2_to_ha = 100

#get land area and cropland area 
total_land_area = country_data_selected['Area (km2)']
total_land_area_km2 = float(total_land_area.iloc[0])  # Convert to a numerical type

cropland_percentage = country_data_selected['Intensity_of_cropland ']
cropland_area_percentage = float(cropland_percentage.iloc[0])  # Convert to a numerical type
cropland_area_km2 = total_land_area_km2 * cropland_area_percentage / 100  # Update cropland_area_km2


# Calculate the total land area and cropland area in hectares
total_land_area_ha = round(total_land_area_km2 * km2_to_ha)
cropland_area_ha = cropland_area_km2 * km2_to_ha

# Obtain the "Barren land" percentage from the selected_crops dictionary
barren_land_percentage = selected_crops.get('Barren land', 0.0)

# Calculate the attainable maximum yield, area cultivated, and other information for each selected crop
total_crop_area_ha = 0
total_maximum_yield = 0

for crop, area_percentage in selected_crops.items():
    # Clean the crop name by stripping leading/trailing spaces
    cleaned_crop_name = crop.strip()
    
    # Calculate the area of land to be cultivated for the selected crop in hectares
    crop_area_percentage = area_percentage
    if cleaned_crop_name != "Barren land":
        crop_area_ha = round((crop_area_percentage / 100) * cropland_area_ha)
    else:
        # For "Barren land," use the provided percentage and calculate the area in hectares
        crop_area_ha = round((barren_land_percentage / 100) * total_land_area_ha)
    
    # Check if the cleaned crop name is in the crop_yield_per_ha dictionary
    if cleaned_crop_name in crop_yield_per_ha:
        # Calculate the attainable maximum yield based on the area and yield per ha for the crop
        yield_per_ha = crop_yield_per_ha[cleaned_crop_name]
        crop_attainable_yield = round(yield_per_ha * crop_area_ha)
        attainable_info[cleaned_crop_name] = {
            'Crop Area (%)': crop_area_percentage,
            'Crop Area (ha)': crop_area_ha,
            'Maximum Yield (tons)': crop_attainable_yield,
        }
        
        # Update total crop area and maximum yield
        total_crop_area_ha += round(crop_area_ha)
        total_maximum_yield += round(crop_attainable_yield)

# Create a table for total land area, total cropland area, maximum yield, and total yield
area_table_data = [
    ["Total Land Area (ha)", total_land_area_ha],
    ["Total Cropland Area (ha)", cropland_area_ha]
]

area_table_headers = ["Area Type", "Value"]
area_table_format = "fancy_grid"

# Display the table for total land area, total cropland area, maximum yield, and total yield
print("\nTotal Land Area, Total Cropland Area, and Total Yield:")
print(tabulate(area_table_data, headers=area_table_headers, tablefmt=area_table_format))

# Display the attainable maximum yield and area information for each selected crop in a fancy table
print("\nAttainable Maximum Yield and Area Information (in hectares):")
table_data = []
total_selected_crop_yield = 0
total_crop_area_percentage = 0
total_crop_area_ha = 0

for crop, info in attainable_info.items():
    table_data.append([crop, f"{info['Crop Area (%)']:.1f}", f"{info['Crop Area (ha)']:.2f} ha", f"{info['Maximum Yield (tons)']:.2f} tons"])
    total_selected_crop_yield += info['Maximum Yield (tons)']
    total_crop_area_percentage += info['Crop Area (%)']
    total_crop_area_ha += info['Crop Area (ha)']

# Add a row for the totals
table_data.append(["Total", f"{total_crop_area_percentage:.1f}", f"{total_crop_area_ha:.2f} ha", f"{total_selected_crop_yield:.2f} tons"])

table_headers = ["Crop", "Crop Area (%)", "Crop Area (ha)", "Maximum Production"]
table_format = "fancy_grid"
print(tabulate(table_data, headers=table_headers, tablefmt=table_format))
```

```python
# Split strategy_data into separate datasets for each group
policy_maker_data = strategy_data[strategy_data['Player '] == 'Policymaker']
community_leader_data = strategy_data[strategy_data['Player '] == 'Community Leader']
farmer_data = strategy_data[strategy_data['Player '] == 'Farmer']

# Function to select strategies for a player
def select_strategies(player_name, player_data):
    player_budget = budgets.get(player_name, 0)
    player_selections = []

    print(f"\n{player_name}'s Budget: {player_budget}")

    while True:
        print(f"\n{player_name}'s Available Strategies:")
        for index, row in player_data.iterrows():
            strategy_no = index + 1  # Start strategy numbering from 1
            strategy = row['Stategy  ']
            price = int(row['Game price'])  # Convert to integer
            print(f"{strategy_no}. {strategy} (Price: {price})")

        selection_no = int(input(f"\n{player_name}, select a strategy by entering its number (or enter 0 to finish): "))
        if selection_no == 0:
            break
        elif 1 <= selection_no <= len(player_data):
            selected_strategy = player_data.iloc[selection_no - 1]
            strategy_price = int(selected_strategy['Game price'])
            if strategy_price <= player_budget:
                player_budget -= strategy_price
                player_selections.append(selected_strategy)
                print(f"{player_name} selected: {selected_strategy['Stategy  ']}")
                print(f"Remaining Budget for {player_name}: {player_budget}")
            else:
                print(f"Insufficient budget. {player_name}'s remaining budget: {player_budget}")
        else:
            print("Invalid selection. Please enter a valid number.")

    return player_selections

# Define budgets for each player
budgets = {'Policymaker': 50, 'Community Leader': 50, 'Farmer': 50}

# Policymaker selects strategies
policy_maker_selections = select_strategies('Policymaker', policy_maker_data)

# Community Leader selects strategies
community_leader_selections = select_strategies('Community Leader', community_leader_data)

# Farmer selects strategies
farmer_selections = select_strategies('Farmer', farmer_data)

def display_selected_strategies(player_name, selected_strategies):
    if selected_strategies:
        table = PrettyTable()
        table.field_names = ["Strategy No.", "Strategy", "Price"]
        total_budget_used = 0
        for index, strategy in enumerate(selected_strategies, 1):
            price = int(strategy['Game price'])  # Convert the price to an integer
            table.add_row([index, strategy['Stategy  '], price])
            total_budget_used += price  # Add the integer price to the total budget used
        remaining_budget = budgets[player_name] - total_budget_used
        budget_used_percentage = (total_budget_used / budgets[player_name]) * 100
        print(f"\n{player_name}'s Selected Strategies:")
        print(table)
        print(f"Total Budget Used: {total_budget_used}, Remaining Budget: {remaining_budget}, Budget Used: {budget_used_percentage:.2f}%")
    else:
        print(f"{player_name} did not select any strategies.")
        
display_selected_strategies('Policymaker', policy_maker_selections)
display_selected_strategies('Community Leader', community_leader_selections)
display_selected_strategies('Farmer', farmer_selections)
```

```python
# Define fun evaluations
fun_evaluations = {
    "Good": "Excellent choices! You've excelled in all aspects.",
    "Fair": "Not bad, but there's room for improvement. Keep it up!",
    "Poor": "Uh-oh! Your choices could use some improvement. Try again!"
}


# Split strategy_data into separate datasets for each group
policy_maker_data = strategy_data[strategy_data['Player '] == 'Policymaker']
community_leader_data = strategy_data[strategy_data['Player '] == 'Community Leader']
farmer_data = strategy_data[strategy_data['Player '] == 'Farmer']

# Function to select strategies for a player

# Calculate scores for each player based on their selected strategies
def calculate_player_scores(player_name, player_selections):
    total_environmental_score = 0
    total_economic_score = 0
    total_social_score = 0

    max_possible_score = len(player_selections) * 100  # Maximum possible score for the player

    for selection in player_selections:
        total_environmental_score += pd.to_numeric(selection['Environment Score '], errors='coerce')  # Convert to numeric
        total_economic_score += pd.to_numeric(selection['Economic Score '], errors='coerce')  # Convert to numeric
        total_social_score += pd.to_numeric(selection['Social Score'], errors='coerce')  # Convert to numeric

    # Calculate the average scores as percentages
    num_selections = len(player_selections)
    if num_selections > 0:
        environmental_score = (total_environmental_score / max_possible_score) * 100
        economic_score = (total_economic_score / max_possible_score) * 100
        social_score = (total_social_score / max_possible_score) * 100
    else:
        # Handle the case when no selections are made
        environmental_score = 0
        economic_score = 0
        social_score = 0

    return {
        'Player': player_name,
        'Environmental Score (Raw)': total_environmental_score,
        'Economic Score (Raw)': total_economic_score,
        'Social Score (Raw)': total_social_score,
        'Environmental Score (%)': environmental_score,
        'Economic Score (%)': economic_score,
        'Social Score (%)': social_score,
        'Total Strategies': num_selections,
        'Maximum Possible Score': max_possible_score
    }

# Calculate scores for each player
policy_maker_scores = calculate_player_scores('Policymaker', policy_maker_selections)
community_leader_scores = calculate_player_scores('Community Leader', community_leader_selections)
farmer_scores = calculate_player_scores('Farmer', farmer_selections)

# Create a table for raw scores and total scores
table_data = [
    ["Player", "Environmental Score (Raw)", "Economic Score (Raw)", "Social Score (Raw)", "Total Strategies", "Maximum Possible Score"],
    [policy_maker_scores['Player'], policy_maker_scores['Environmental Score (Raw)'], policy_maker_scores['Economic Score (Raw)'], policy_maker_scores['Social Score (Raw)'], policy_maker_scores['Total Strategies'], policy_maker_scores['Maximum Possible Score']],
    [community_leader_scores['Player'], community_leader_scores['Environmental Score (Raw)'], community_leader_scores['Economic Score (Raw)'], community_leader_scores['Social Score (Raw)'], community_leader_scores['Total Strategies'], community_leader_scores['Maximum Possible Score']],
    [farmer_scores['Player'], farmer_scores['Environmental Score (Raw)'], farmer_scores['Economic Score (Raw)'], farmer_scores['Social Score (Raw)'], farmer_scores['Total Strategies'], farmer_scores['Maximum Possible Score']]
]

# Print the scores in a table format
score_table = tabulate(table_data, headers="firstrow", tablefmt="fancy_grid")
print("\nRaw and Percentage Environmental, Economic, and Social Scores:")
print(score_table)

# Print the total scores in a table format
total_scores_data = [
    ["Player", "Total Environmental Score (%)", "Total Economic Score (%)", "Total Social Score (%)"],
    [policy_maker_scores['Player'], policy_maker_scores['Environmental Score (%)'], policy_maker_scores['Economic Score (%)'], policy_maker_scores['Social Score (%)']],
    [community_leader_scores['Player'], community_leader_scores['Environmental Score (%)'], community_leader_scores['Economic Score (%)'], community_leader_scores['Social Score (%)']],
    [farmer_scores['Player'], farmer_scores['Environmental Score (%)'], farmer_scores['Economic Score (%)'], farmer_scores['Social Score (%)']]
]

# Print the total scores in a table format
total_score_table = tabulate(total_scores_data, headers="firstrow", tablefmt="fancy_grid")
print("\nTotal Scores for Each Player:")
print(total_score_table)

# Calculate the EES (Environmental, Economic, Social) score as an average
average_environmental_score = (policy_maker_scores['Environmental Score (%)'] + community_leader_scores['Environmental Score (%)'] + farmer_scores['Environmental Score (%)']) / 3
average_economic_score = (policy_maker_scores['Economic Score (%)'] + community_leader_scores['Economic Score (%)'] + farmer_scores['Economic Score (%)']) / 3
average_social_score = (policy_maker_scores['Social Score (%)'] + community_leader_scores['Social Score (%)'] + farmer_scores['Social Score (%)']) / 3

Overall_EES_score = (average_environmental_score + average_economic_score + average_social_score )/3

# Print the average scores
print(f"\nAverage Environmental Score: {average_environmental_score:.2f}")
print(f"Average Economic Score: {average_economic_score:.2f}")
print(f"Average Social Score: {average_social_score:.2f}")

# Define evaluation thresholds (you can adjust these as needed)
good_threshold = 70
fair_threshold = 50

# Evaluation function
def evaluate_player(player_scores):
    if player_scores['Environmental Score (%)'] >= good_threshold and player_scores['Economic Score (%)'] >= good_threshold and player_scores['Social Score (%)'] >= good_threshold:
        return "Good"
    elif player_scores['Environmental Score (%)'] >= fair_threshold and player_scores['Economic Score (%)'] >= fair_threshold and player_scores['Social Score (%)'] >= fair_threshold:
        return "Fair"
    else:
        return "Poor"

print(f"Overall EES_Score: {Overall_EES_score}")
# Evaluate each player
policy_maker_evaluation = evaluate_player(policy_maker_scores)
community_leader_evaluation = evaluate_player(community_leader_scores)
farmer_evaluation = evaluate_player(farmer_scores)

# Print player evaluations with fun phrases
print("\nPlayer Evaluations:")
print(f"Policymaker: {policy_maker_evaluation} - {fun_evaluations.get(policy_maker_evaluation)}")
print(f"Community Leader: {community_leader_evaluation} - {fun_evaluations.get(community_leader_evaluation)}")
print(f"Farmer: {farmer_evaluation} - {fun_evaluations.get(farmer_evaluation)}")
```

```python
#Calculate corporation score 
# Initialize alignment scores as variables
alignment_score_policymaker_community_leader = 0
alignment_score_policymaker_farmer = 0
alignment_score_community_leader_farmer = 0
alignment_score_policymaker_community_leader_farmer = 0
alignment_score_all_players = 0  # Initialize the new variable for all 3 players

def calculate_alignment_score(player1_name, player1_selections, player2_name, player2_selections):
    common_selections = set(selection1['No.'] for selection1 in player1_selections).intersection(
        selection2['No.'] for selection2 in player2_selections
    )

    alignment_score = len(common_selections)

    print(f"{player1_name} vs {player2_name}: Alignment Score: {alignment_score} common selections")

    # Update alignment scores as numeric variables
    if player1_name == 'Policymaker' and player2_name == 'Community Leader':
        global alignment_score_policymaker_community_leader
        alignment_score_policymaker_community_leader = alignment_score
    elif player1_name == 'Policymaker' and player2_name == 'Farmer':
        global alignment_score_policymaker_farmer
        alignment_score_policymaker_farmer = alignment_score
    elif player1_name == 'Community Leader' and player2_name == 'Farmer':
        global alignment_score_community_leader_farmer
        alignment_score_community_leader_farmer = alignment_score
    elif player1_name == 'Policymaker' and player2_name == 'Community Leader and Farmer':
        global alignment_score_policymaker_community_leader_farmer
        alignment_score_policymaker_community_leader_farmer = alignment_score
    elif player1_name == 'Farmer' and player2_name == 'Policymaker and Community Leader':
        global alignment_score_all_players
        alignment_score_all_players = alignment_score

# Calculate alignment between all three players
calculate_alignment_score('Policymaker', policy_maker_selections, 'Community Leader', community_leader_selections)
calculate_alignment_score('Policymaker', policy_maker_selections, 'Farmer', farmer_selections)
calculate_alignment_score('Community Leader', community_leader_selections, 'Farmer', farmer_selections)
calculate_alignment_score('Policymaker', policy_maker_selections, 'Community Leader and Farmer', community_leader_selections + farmer_selections)
calculate_alignment_score('Farmer', farmer_selections, 'Policymaker and Community Leader', policy_maker_selections + community_leader_selections)

# Now, you have alignment scores as the count of common selections between players.

# Calculate the maximum alignment score
max_alignment_score = max(alignment_score_policymaker_community_leader,
                          alignment_score_policymaker_farmer,
                          alignment_score_community_leader_farmer,
                          alignment_score_policymaker_community_leader_farmer,
                          alignment_score_all_players)

# Normalize alignment scores
alignment_score_policymaker_community_leader /= max_alignment_score
alignment_score_policymaker_farmer /= max_alignment_score
alignment_score_community_leader_farmer /= max_alignment_score
alignment_score_policymaker_community_leader_farmer /= max_alignment_score
alignment_score_all_players /= max_alignment_score

# Define alignment score thresholds
good_threshold = 0.7
bad_threshold = 0.4

# Function to classify alignment
def classify_alignment(alignment_score):
    if alignment_score >= good_threshold:
        return "Good"
    elif alignment_score <= bad_threshold:
        return "Bad"
    else:
        return "Fair"

# Add alignment classification to the DataFrame
alignment_scores_table = pd.DataFrame({
    'Player 1': ['Policymaker', 'Policymaker', 'Community Leader', 'Policymaker', 'Farmer'],
    'Player 2': ['Community Leader', 'Farmer', 'Farmer', 'Community Leader and Farmer', 'Policymaker and Community Leader'],
    'Alignment': [
        alignment_score_policymaker_community_leader,
        alignment_score_policymaker_farmer,
        alignment_score_community_leader_farmer,
        alignment_score_policymaker_community_leader_farmer,
        alignment_score_all_players
    ]
})

alignment_scores_table['Alignment Rating'] = alignment_scores_table['Alignment'].apply(classify_alignment)

print("\nCorporation Scores and Rating:")
print(alignment_scores_table)

# Calculate the overall alignment score as the average alignment of all players
overall_alignment_score = (alignment_score_policymaker_community_leader +
                           alignment_score_policymaker_farmer +
                           alignment_score_community_leader_farmer +
                           alignment_score_policymaker_community_leader_farmer +
                           alignment_score_all_players) / 5

print(f"\nOverall Corporation Score: {overall_alignment_score:.2f}")

# Classify the Overall Alignment Score based on the existing thresholds
overall_alignment_classification = classify_alignment(overall_alignment_score)
print(f"Overall Corporation Rating: {overall_alignment_classification}")


# Save player selections to CSV files
policy_maker_selections_df = pd.DataFrame(policy_maker_selections)
policy_maker_selections_df.to_csv('policy_maker_selections.csv', index=False)

community_leader_selections_df = pd.DataFrame(community_leader_selections)
community_leader_selections_df.to_csv('community_leader_selections.csv', index=False)

farmer_selections_df = pd.DataFrame(farmer_selections)
farmer_selections_df.to_csv('farmer_selections.csv', index=False)

print("\nPlayer selections saved to CSV files: policy_maker_selections.csv, community_leader_selections.csv, farmer_selections.csv")
```
